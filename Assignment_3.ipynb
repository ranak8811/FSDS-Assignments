## 1. Inventory Price Calculator

```python
def calculate_inventory_summary (prices):
  total_cost = sum(prices)
  average_cost = sum(prices) / len(prices)

  return total_cost, average_cost

prices_str = input("Enter prices separated by commas (e.g., 120,250,399): ")
l = [float(p.strip()) for p in prices_str.split(',')]

print(calculate_inventory_summary(l))
```

## 2. Unique Visitor Tracker

```python
def count_unique_visitors(visitor_list):
  unique_visitors = set(visitor_list)

  no_unique_visitors = len(unique_visitors)
  return no_unique_visitors, unique_visitors

visitors_str = input("Enter visitor IDs separated by commas (e.g., 101,205,101): ")
id_l = [int(v.strip()) for v in visitors_str.split(',')]

no_unique_visitors, unique_visitors = count_unique_visitors(id_l)

print(f"{no_unique_visitors} (The unique IDs are {', '.join(str(id) for id in unique_visitors)})")
```

## 3. Product Category Counter

```python
def count_categories(products):
  occurrence_dict = dict()
  for p in products:
    if p not in occurrence_dict.keys():
      occurrence_dict[p] = 1
    else:
      occurrence_dict[p] = occurrence_dict[p] + 1

  return occurrence_dict

products_str = input("Enter product categories separated by commas (e.g., Laptop,Smartphone,Monitor): ")
product_l = [p.strip() for p in products_str.split(',')]

print(count_categories(product_l))
```

## 4. Temperature Analytics

```python
def analyze_temperatures(temp_list):
  high_list = []
  normal_temp = []

  for temp in temp_list:
    if temp > 38:
      high_list.append(temp)
    else:
      normal_temp.append(temp)

  return high_list, normal_temp

temps_str = input("Enter temperatures separated by commas (e.g., 36.5,38.0,39.1): ")
temperature_l = [float(t.strip()) for t in temps_str.split(',')]

result = analyze_temperatures(temperature_l)

high_list, normal_list = result

print(f"High List (>38째C): {high_list}")
print(f"Normal List (<38째C): {normal_list}")

print(f"Full Output Tuple: {result}")
```

## 5. Menu Price Lookup

```python
def get_price(menu, item):
  if item in menu.keys():
    return menu[item]
  else:
    return "Item not found"

Menu = {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}
item_name = input("Enter the item name to look up (e.g., Muffin): ")

print(get_price(Menu, item_name))
```

## 6. Student Grading System

```python
def calculate_grade(score):
  if score >= 80:
    return 'A+'
  elif 70 <= score <= 79:
    return 'A'
  elif 60 <= score <= 69:
    return 'A-'
  elif 50 <= score <= 59:
    return 'B'
  else:
    return 'F'

def grade_students(student_dict):
  for k, v in student_dict.items():
    student_dict[k] = calculate_grade(v)

  return student_dict

marks_dict = {}
try:
  num_students = int(input("Enter the number of students to grade: "))
  for i in range(num_students):
      name = input(f"Enter name for student {i+1}: ")
      score = int(input(f"Enter score for {name}: "))
      marks_dict[name] = score
except ValueError:
  print("Invalid input. Please enter a valid number for count and score.")

print(grade_students(marks_dict))
```

## 7. Discount Calculator

```python
def apply_discounts(products):
  return [p[1] - (p[1]*(p[2]/100)) for p in products]

products = []
try:
  num_products = int(input("Enter the number of products: "))
  for i in range(num_products):
      name = input(f"Enter name for product {i+1}: ")
      price = float(input(f"Enter price for {name}: "))
      discount = float(input(f"Enter discount percentage for {name}: "))
      products.append((name, price, discount))
except ValueError:
  print("Invalid input. Please enter valid numbers for price and discount.")

print(apply_discounts(products))
```

## 8. Hashtag Generator

```python
def generate_hashtags(sentence):
  return [f'#{w.lower()}' for w in sentence.split()]

sentence = input("Enter a sentence to generate hashtags (e.g., AI Data Science): ")
print(generate_hashtags(sentence))
```

## 10. Shopping Cart Bill Calculator

```python
def calculate_total(cart, prices):
  sum = 0

  for c in cart:
    if c in prices:
      sum += prices[c]
    else:
      print(f"Warning: Item '{c}' not found in price list. Skipping.")
  return sum

cart_items_str = input("Enter items in the cart separated by commas (e.g., Milk,Bread,Milk): ")
cart = [item.strip() for item in cart_items_str.split(',')]
prices = {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00}

print(calculate_total(cart, prices))
```
