## Assignment 1: Temperature Check

```python
temp = int(input())

if temp < 10:
  print("It's too cold!")
elif 10 <= temp <= 25:
  print("It's cool outside")
else:
  print("It's hot outside!")
```

## Assignment 2: ATM Withdrawal System

```python
balance = int(input('Enter your account balance: '))
withdraw_amount = int(input('Enter the amount you want to withdraw: '))

if withdraw_amount > balance:
  print("Insufficient Balance")
elif withdraw_amount <= balance:
  print("Transaction Successful")
  print("Remaining balance: ", balance - withdraw_amount)
```

## Assignment 3: Student Grade Calculator

```python
marks = int(input())

if 90 <= marks <= 100:
  print('A+')
elif 80 <= marks <= 89:
  print('A')
elif 70 <= marks <= 79:
  print('B')
elif 60 <= marks <= 69:
  print('C')
else:
  print('Fail')
```

## Assignment 4: Bus Ticket Discount

```python
age = int(input())

price = 100

if age < 5:
  print('Ticket price for you: ', 0)
elif 5 <= age <= 18:
  print('Ticket price for you: ', price - price/2)
elif age >= 60:
  print('Ticket price for you: ', price - price*(30/100))
else:
  print('Ticket price for you: ', price)
```

## Assignment 5: Restaurant Bill with Discount

```python
bill = int(input())

if bill > 1000:
  print('Amount to pay: ', bill - bill * (10/100))
else:
  print('Amount to pay: ', bill)
```

## Assignment 7: Traffic Light System

```python
color = input('Enter traffic light color name: ')

if color == 'red':
  print('Stop')
elif color == 'yellow':
  print('Ready to go')
elif color == 'green':
  print('Go')
else:
  print('Invalid color')
```

## Assignment 8: Leap Year Checker

```python
year = int(input())

if year % 400 == 0:
  print('Leap Year')
elif year % 4 == 0 and year % 100 != 0:
  print('Leap Year')
else:
  print('Not a Leap Year')
```
