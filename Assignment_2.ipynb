## 1. Email Validator (String Filtering + Loops)

```python
email = input('Enter your email: ')

prefix, domain = email.split('@')

if (prefix and (domain.endswith('.com'), domain.endswith('.org'))):
  print('Valid Email')
else:
  print('Unvalid Email')
```

## 2. Count Word Frequency in Customer Reviews

```python
text = input('Enter any text: ').lower()

freq = {}

for w in text.split():
  if w in freq.keys():
    freq[w] = freq[w] + 1
  else:
    freq[w] = 1

print(freq)
```

## 3. Password Strength Checker

```python
password = input('Enter your password: ')

special_characters = "!@#$%^&*()-+?_=,<>/"

uc = 0; lc = 0; dc = 0; sc = 0;

for l in password:
  if l.isupper():
    uc += 1
  elif l.islower():
    lc += 1
  elif l.isdigit():
    dc += 1
  elif l in special_characters:
    sc += 1

if uc > 0 and lc > 0 and dc > 0 and sc > 0:
  print('Strong Password')
else:
  print('Weak Password')
```

## 4. Username Generator

```python
import random

f_name = input('Enter your full name: ')

u_name = ''

first_part_c = 3
last_part_c = len(f_name) - 4

for i in range(len(f_name)):
  if i < first_part_c:
    u_name += f_name[i]
  elif i > last_part_c:
    u_name += f_name[i]

u_name = u_name + str(random.randint(1, 100))

print(u_name)
```

## 5. SMS Word Counter

```python
sms = input('Enter your SMS: ')

char_c = len(sms)

word_c = len(sms.split())

if char_c > 160:
  print(f"Words = {word_c}, Characters = {char_c}, Status = Beyond Limit")
else:
  print(f"Words = {word_c}, Characters = {char_c}, Status = Within Limit")
```

## 6. Remove Duplicates from Contact List

```python
names = input('List of names: ')

names_list = names.split(',')

unique_names = []

for name in names_list:
  if name not in unique_names:
    unique_names.append(name)

unique_names = ', '.join(unique_names)

print(unique_names)
```

## 7. Detect Suspicious Words in Text (Moderation)

```python
text = input('Enter any text: ')

banned_w = ['spam', 'scam', 'fake']

suspicious_detected = False

for word in text.split():
  if word in banned_w:
    suspicious_detected = True
    break

if suspicious_detected:
  print('Warning: Suspicious content detected')
else:
  print('No Suspicious content detected')
```

## 9. Count Hashtags and Mentions (Social Media Analytics)

```python
post = input('Enter any post: ')

hashtag_count = post.count('#')
mention_count = post.count('@')

print(f"Hashtags = {hashtag_count}, Mentions = {mention_count}")
```

## 10. Word Reversal for Data Encryption

```python
text = input('Enter any text: ')

reversed_text_list = [w[::-1] for w in text.split()]

print(' '.join(reversed_text_list))
```
